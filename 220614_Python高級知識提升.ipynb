```python
《Python允許在類外部增加物件屬性，但不推薦 - "物件.屬性=值"》
https://www.udemy.com/course/python-o/learn/lecture/15642472#overview


# 不推薦原因：屬性應該被封裝在類的內部


# Python解釋器看到class之後，會記得有ClassName這個類，但不會跳進去執行裏面的code。這和看到def的情況是一樣的
class ClassName(object):
    """類說明"""
    # 有self，所以是物件方法，也就是對物件做初始化
    def __init__(self, arg):
        # 自xxx類繼承物件初始化
        super(ClassName, self).__init__() 
        self.arg = arg # 定義自己的物件屬性

    def eat(self): # self: <__main__.ClassName object at 0x7f53495524e0>
        print("eat")


tom = ClassName() 
# 使用pycharm驗證，Python允許在類外部增加物件屬性。在「tom.name = "Tommy"」這邊設斷點
tom.name = "Tommy" # tom={ClassName}<__main__.ClassName object at 0x7f53495524e0>
# 會得到tom是{ClassName}類，值ClassName object，記憶體地址0x7f53495524e0
# tom的下拉會出現「arg={NoneType}None」
# 也就是tom有一個物件屬性arg，是{NoneType}類，值None
# 當點擊"步入 Step Into"，tom的下拉就會多出現「name={str}'Tommy'」
# 這就是Python允許在類外部增加物件屬性的證明
tom.eat() # 點擊"步入 Step Into"，調用self.eat()方法
# 會出現self={ClassName} <__main__.ClassName object at 0x7f53495524e0>
# 可以觀察到，self和tom指向的地址都是0x7f53495524e0。代表，「self這個物件，就是tom這個物件」！！
# 哪一個物件調用方法，self就是指向那一個物件的地址
# self的下拉會出現arg={NoneType}None，以及name={str}'Tommy'

```

```python
《魔法(內建)方法》
# 構造方法
# python中的__new__創建方法，和__init__初始化方法合起來，就是其他語言中的「構造方法」


__del__()
# 物件在準備從記憶體銷毀之前，會先自動調用__del__方法
# 也就是說，比如程序要結束前，所有的物件都會從記憶體釋放，那這時候物件們就會各自自動調用__del__方法
# 又或者是，在code中，自己主動寫"del 物件"，那也會調用__del__方法
# 所以比如想檢查記憶體有沒有被釋放，可以自定義__del__方法，去print一段訊息來識別


__str__()
# print(物件)的預設輸出是 -> 物件類名 object at 記憶體位置(16進制)
# 在類中自定義__str__方法，就可以改變print(物件)的預設輸出了。因爲是print，所以__str__必須返回一個字符串


__call__()
# 物件名()，會觸發執行物件的__call__方法
# "xx()" 可以理解為是 "xx.__call__()" 的簡寫
class Foo:
    def __init__(self):
        print("--執行init方法--")

    def __call__(self, *args, **kwargs):
        print("--執行call方法--")

foo = Foo() # --執行init方法--
foo() # --執行call方法--


__getitem__()、__setitem__()、__delitem__()
# iterable物件，是因爲實現了iter()和next()
# 同理，只要實現了__getitem__()、__setitem__()、__delitem__()，就也可以把物件當作字典用
class Foo:

    def __getitem__(self, key):
        print('__getitem__, key=%s' %key)

    def __setitem__(self, key, value):
        print('__setitem__, key=%s, value=%s' %(key, value))

    def __delitem__(self, key):
        print('__detitem__, key=%s' %key)

foo = Foo()
# 觸發調用__getitem__
foo['kkk'] # __getitem__, key=kkk
# 觸發__setitem__
foo['kkk'] = 'vvv' # __setitem__, key=kkk, value=vvv
# 觸發__delitem__
del foo['kkk'] # __detitem__, key=kkk

```

```python
《__getslice__ - 2.x版本》
# __getslice__()等方法用于切片操作，仅適用於2.x版本
# 這邊也是實現了__getslice__()等方法，就可以把物件當作列表操作
# 就如同上面的實現__getitem__()等方法，就可以把物件當作字典操作
class Foo(object):
    def __getslice__(self, i, j):
        print('__getslice__', i, j)

    def __setslice__(self, i, j, sequence):
        print('__setslice__', i, j,sequence)

    def __delslice__(self, i, j):
        print('__delslice__', i, j)


obj = Foo()
# 自动触发执行__getslice__
obj[-1:1]
# 自动触发执行__setslice__
obj[0:1] = [11, 22, 33, 44]
# 自动触发执行__delslice__
del obj[0:2]
```

```python
# 《魔法方法 - 切片》
# https://www.jianshu.com/p/1af77113d70b

# Python3中的切片的魔术方法不再是Python2中的getslice(), setslice()和delslice()
# 而是借助 slice 类整合到了getitem(),setitem()和 delitem()中。


class Foo(object):
    def __getitem__(self, index):
        if isinstance(index, slice): # obj[-1:10] 是 slice 類
            # 不知道index.start該用%d還是什麼的話，把自己轉成%s就可以被print出來了
            print("getitem> start: %s, stop: %s, step: %s." % (str(index.start), str(index.stop), str(index.step)))
            print(index) 

    def __setitem__(self, index, value):
        if isinstance(index, slice):
            print("setitem> start: %s, stop: %s, step: %s." % (str(index.start), str(index.stop), str(index.step)))
            print("The value is:", value)

    def __delitem__(self, index):
        if isinstance(index, slice):
            print("delitem> start: %s, stop: %s, step: %s." % (str(index.start), str(index.stop), str(index.step)))


if __name__ == "__main__":
    obj = Foo()
    obj[-1:10] 
    # getitem> start: -1, stop: 10, step: None.
    # slice(-1, 10, None)
    obj[-1:10:1] = [2, 3, 4, 5] 
    # setitem> start: -1, stop: 10, step: 1.
    # The value is: [2, 3, 4, 5]
    del obj[-1:10:2]
    # delitem> start: -1, stop: 10, step: 2.
```

```python
《super》
# 如果直接同名同參數改寫(overwrite)父類方法，那就是廢棄了父類，完全只依照子類的code去做
# 如果要引用父類方法當作子類中的一部分code，那就要寫super().父類方法
# super是一個很特別的類，所以super()是super類創建出來的物件
# 比如父類是Dog，子類是Xiaotian，都有bark()方法。那子類要調用父類，就是創建super物件去掉用，super().bark()
# Python2.7版本，沒有super，所以要寫成「父類名.方法(selfs)」，變成Dog.bark(self)。缺點是，子類code中所有要調用父類方法的地方，全部都要替換
# 如果寫錯了，比如Xiaotian類中要調用Dog.bark(self)，但寫成自己的Xiaotian.bark(self)，方法內部調用自己正在寫的方法，那就會是死循環而不停報錯同一行
# 父類的私有屬性、方法，子類是看不到的
# 要調用父類私有屬性、方法，要嘛前面加_類名，要嘛父類的公有方法中，要有提供私有屬性、方法的調用API，那子類的物件就可以透過API去調用到父類的私有屬性、方法了


《繼承》
# python3中都是新式類，如果不給參數可以連()都省略，也會自動繼承object。相當於Python2中的「class 類名(object):」。
# 既然要自動繼承，就代表有好處，那就是擁有各種魔法(內建)方法
# 經典(舊式)類中，dir(物件名)查找所有魔法(內建)方法，只會含有['__doc__', '__module__']
# dir(物件名)不是存在於類中的魔法(內建)方法，所以呼叫不用加底線


《多繼承》
# 如果父類之間存在同名屬性/方法，那就應該盡量避免使用多繼承
# 如果還是使用了多繼承，那就會依照解析順序（Method Resolution Order, MRO）的繼承先後順序，去選擇調用哪個類下的方法
# 類有一個魔法(內建)屬性__mro__，裏面存放了調用的路徑

```

```python
《靜態方法》
# 在pycharm中，如果發現封裝在類中的方法，是既不需要訪問實例屬性，也不需要訪問類屬性，那就會提示可能是「靜態方法」
# 雖然實質上是個靜態方法的體質，但如果把self硬是加進去當第一個參數，也只是提示可能是靜態方法，但執行起來是不會報錯的
# 如果刪除掉self，則pycharm會有虛線提示，通常第一個參數是self，因爲一個沒掛類或靜態裝飾器的方法，會被認定就是實例方法
# 所以爲了一切沒有問題提示的虛線會再次產生了，這時候要明確加上@staticmethod
# 類方法加上@classmethod也是同理，並且裝飾器也可以增加可閱讀性
# 實例方法因爲最常見，所以Python設計其不需要加@
# 因爲靜態方法是封裝在類中，所以要調用的話，可以使用「類名.靜態方法名」
# 結論，要訪問物件屬性的就用物件方法，訪問類屬性的是類方法。2者屬性都不訪問，但要封裝在類中的方法，就是靜態方法
# 如果既要訪問實例屬性，又要訪問類屬性，那就用實例方法。因爲在實例方法的內部中，可以使用「類名.類屬性」去訪問到類屬性
# 而如果在類的內部中，想要去訪問到物件屬性，就必須使用到self，那就還是會變成是實例方法了

```

```python
《內建方法 - __new__》
# 類名()指令，會調用__new__()，在記憶體中取得一塊分配到的空間
# 並返回物件地址(或稱爲引用)，讓__init__()中的self去指向這個地址
# 這也是創建完物件後，去print(物件)，會返回物件地址的原因。因爲就是__init__會從__new__那邊取得到，物件所分配到的記憶體起始地址
# 所謂魔法方法，其實就是源自父類(基類)object的內建方法
# 重寫__new__方法，一定要「return super().__new__(cls)」，這樣才會有去向記憶體取得空間，以及傳送起始地址給要調用的__init__
# 創建super物件，去調用父類的__new__()方法
# __new__是一个静态方法，而__init__是一个实例方法
# __new__方法会返回一个创建的实例，而__init__什么都不返回
# object.__new__(cls[, ...])调用以创建一个 cls 类的新实例


# 創建物件時，首先會自動調用__new__()
# 因爲__new__()是靜態方法，所以cls並不是強迫在首參數，而是一般參數，代表目前調用的類
def __new__(cls, *args, **kwargs):
    # 使用super()去取得父類的__new__方法，才有辦法去向記憶體取得空間，以及傳送起始地址給要調用的__init__
    # 這個cls是代表從def __new__中取得到的實參，傳遞給父類的__new__方法去建構一連串的記憶體取得、調用__init__等程序
    instance = super().__new__(cls)
    # 不返回物件(引用)地址，解釋器就不會調用__init__方法
    return instance

```

```python
《为什么__new__中的cls能表示当前类？》
https://www.zhihu.com/question/318097303

首先，staticmethod并不强制要求第一个参数为当前类（可以是任何普通参数），
但是这并不妨碍我们把当前类就当作一个普通参数作为staticmethod的参数，
也可以作为第二个、第三个参数（可以作为第一个参数，像__new__）。
cls只是一个变量名，并不代表什么。重要是函数被call的时候，是用的什么参数值。

其次，为什么__new__这个staticmethod被call的时候，第一个参数的参数值就是类呢？
这是因为__new__作为一个python的magic method，是被called来创造类的对像的；
它的返回值就是类的对象。要创造一个类的对象，就必须需要当前类的信息，
所以当前类就被__new__当作第一个参数的参数值。

然后，既然__new__被called的时候第一个参数的参数值为类，为什么不把__new__设计成为classmethod呢？
因为__new__的第一个参数可以不等于当前类（而是其他类，比如base类）。
如果把__new__ 设计为classmethod，那么__new__第一个参数必须为当前类。
但是有的时候，比如有继承结构的时候，我们要在base类里调用__new__来创造子类的对象。
这个时候，如果__new__ 是classmethod就做不到了。

最后，把类class当作staticmethod的参数之一其实是factory design pattern里常用到的。
比如在base类里定义一个staticmethod，第一个参数为某个类。
然后可以call这个staticmethod创造出很多不同类的对象（比如很多不同子类的对象）。

```

```python
《單例-在系統中只有一個實例》
# 「設計模式-單例」，讓類建立的物件，在系統中只有一個實例，比如：音樂播放同時間只有一首，資源回收桶一個操作系統只有一個，印表機同一時間只印一個資料來源
# 設計模式，針對某一特定問題，前人成熟的解決模式
# 「類名()」在系統中，也是單例，也就是不管執行幾次，返回的記憶體地址都是同一個
# 也就是說，記憶體中的類就只有一份，但可以做出多個物件


class Single_ins(object):
    """讓這個類在系統中只能創建一次單例"""
    # 用來記錄實例是不是已經被創建過了
    instance = None
    # 用來記錄是不是執行過初始化的動作了
    init_flag = False

    # 類方法的cls，是首要參數，用於指定當前類
    # __new__是靜態方法，所以cls是一般參數，通常也是指定當前類沒錯
    # 但有繼承關系的時候，卻不一定是指定當前類
    def __new__(cls, *args, **kwargs):
        # 先看看目前有沒有實例，有的話就直接返回，沒的話才進去if做創建
        # None是一個特別物件，而「物件用is」
        if instance is None:
            # super是一個特殊類，所以要建立物件，才能去調用物件方法
            # 不能直接「return super().__new__(cls)」
            # 因爲這樣就不會去改變instance的記錄值
            # 使用當前類作爲參數，去呼叫MRO中排序第一的父類__new__方法
            # 然後讓當前類屬性instance去指向，返回的記憶體地址
            # 所以此時類屬性instance，就不再會是None了
            cls.instance = super().__new__(cls)         
        return cls.instance

    # 每次創建物件時，除了__new__以外，還會調用__init__
    # 雖然每次創建物件時，已經讓__new__不再創建新的物件（只回傳同一個地址）
    # 但每次__new__去調用__init__時，這個物件都會不斷的被重新初始化
    # 如果是不想讓物件不斷的被重新初始化，那就要執行下面的動作    
    def __init__(self):
       if Single_ins.init_flag:
        # 如果已經執行過初始化了，那就不需要執行下方的code
        # 使用return，就可以直接結束這個方法
        # return後面默認是None
        return

        # 如果沒有初始化過，那就要執行初始化動作，並且把flag做更改
        print("執行初始化動作")
        Single_ins.init_flag = True 
```

```python
《exception》
# 有主動避免結束程序的try - except Exception，那相對也有主動拋出異常，結束程序的raise - Exception("xxx")
# Exception是一個類，所以可以建立物件。而raise是關鍵字，後面接上Exception類的物件
# 智能提示會顯示，Exception(*args)，代表裏面可以放不定參數


def foo():
    ex = Exception("錯誤提示訊息") # 創建Exception類的物件
    raise ex # raise是關鍵字，後面接上Exception類的物件

try:
    foo() # 接收到拋出的異常
except Exception as e: # 拋出的異常類型，符合這個except中可以被捕捉到類型，所以會執行下面巢狀中的code
    # 在類中自定義__str__方法，就可以改變print(物件)的預設輸出了。因爲是print，所以__str__必須返回一個字符串
    # 而Exception類的__str__方法，會去返回Exception()中的首參數字符串
    print(e) # 印出「錯誤提示訊息」
```

```python
# 《logging - 輸出到控制臺&文件》

# logging的好處是，可以同時print在銀幕上，並儲存在file中

import logging

# 創建日志物件logger，「記錄log」
# 不創建的話，默認會是root
logger = logging.getLogger()
logger.setLevel(logging.INFO) # 輸出等級的「總開關」


# 創建handler物件fh，「操作輸出到file」
logfile = './log.txt' # 在當前目錄下建一個log.txt檔寫入日志
fh = logging.FileHandler(logfile, mode='a') # log應該就是要不斷的append進去，而不是w的洗掉重寫
fh.setLevel(logging.DEBUG) # 輸出到file的等級開關


# 創建handler物件ch，「操作輸出到控制臺」
ch = logging.StreamHandler()
ch.setLevel(logging.WARNING) # 輸出到console控制臺的等級開關


# 對handler設置輸出格式
# 創建Formatter物件，參數爲輸出的格式
formatter = logging.Formatter("%(asctime)s - %(filename)s[line:%(lineno)d] - %(levelname)s: %(message)s")
fh.setFormatter(formatter)
ch.setFormatter(formatter)


# 讓「記錄log」和「操作輸出的handler」進行關聯
logger.addHandler(fh)
logger.addHandler(ch)


# 開始記錄日志
# 實際的log應用，比如網管可以去記錄，有誰在什麼時間訪問了什麼檔案，或開啓了什麼url之類的，多久以後才關閉url之類的
# 或是記錄下那些，發生try-except前後，或是什麼條件出發不符合if的情況，以作爲後續解決問題的復現資訊
# 或是要統計分析數據時，也需要很多記錄，比如user的點擊路徑行爲，是從哪個頁面一直持續到哪個頁面
# 哪個url被點擊的多，到達哪個網頁時又會導致user離開率最高，之類的，都可以作爲記錄保存下來，以作後續分析
logger.warning("this is logging WARNING message")
# 2022-06-06 02:05:46,407 - <ipython-input-11-7830ec282369>[line:37] - WARNING: this is logging WARNING message
```

```python
《type() - 元類》
# Python最底層的，就是元類。tuple、list、dict、class、function等物件，都是由元類所創建出來的物件
# 類，是一個模板，用來創建實例物件


# 動態創建類，傳入的實參是'foo'就返回Foo類
# 其他實參就返回Bar類
def which_class(class_name):
    if class_name == 'foo':
        class Foo(object):
            pass
        return Foo # 返回的是類，不是類的實例
    else:
        class Bar(object):
            pass
        return Bar

my_class = which_class('foo')
print(my_class) # 返回的是類，不是類的實例
# <class '__main__'.Foo>
print(my_class()) # 創建類實例，也就是物件
# <__main__.Foo object at 0x89c6d4c>


# ObjectCreator的type，居然是'type'！？
print(type(1)) # <type 'int'>
print(type(ObjectCreator)) # <type 'type'>


# 一個函數根據傳入的參數不同，可以有2總以上完全不同的功能
# 這時候閱讀性上容易造成混淆，所以通常一個函數功能就應該單純的只有一個
# 但type因爲歷史因素，爲了相容性，保留了2個功能
# 當type傳入3個參數時，功能是創建一個類
# type(類名, 父類名稱組成的元組(根據繼承情況，可以爲空), 包含屬性的字典(名稱會值))


# 使用help去查看，可以發現type不是一個函數，而是一個類，而且有2種使用方式
help(type)
# class type(object)
#  |  type(object) -> the object's type
#  |  type(name, bases, dict) -> a new type

```

```python
《type - metaclass》
# python的類，是根據metaclass來創建的，而metaclass的default值是type
# 也就是說，我們用type來創建類，其實只是預設，要創建類還繼續有花式可以玩
# 函數type_cls的返回必須是個類物件，才能當作metaclass的賦值，用於去創建新的類
# 既然函數的返回也必須是個類物件，那不就和type類似嗎，所以我們就先以和type一樣的參數，給函數type_cls()代入作爲開頭練習

def type_cls(class_name, class_parents, class_attr):
    return type(class_name, class_parents, class_attr)

class Foo(object, metaclass=type_cls):
    attr1 = 'bar'

# hasattr(類名, '屬性名')
print(hasattr(Foo, 'attr1')) # True
print(hasattr(Foo, 'ATTR1')) # False

f = Foo()
print(f.attr1) # bar

------------------------

# python2，指定type_cls的寫法
class Foo(object):
    __metaclass__ = type_cls

```

```python
《type - metaclass應用2》
# metaclass可以對類做修改，而不改動類裏面的code
# 這就和裝飾器的功能很像，只不過@decorator是對func做修改，metaclass是對class做修改


# 既然利用自定義metaclass，類名、繼承、類屬性都應該可以做修改
# 那Python3創建類都會自動繼承自object，也應該可以藉由自定義metaclass，去對一些自定義的類，修改成不從object做繼承
# 比如在type(class_name, class_parents, class_attr)中，去遍歷class_parents這個元組
# 如果有object，就做一個沒有object的新元組，然後回傳這個metaclass


# 直覺上來說，會覺得metaclass是創建類，應該是個class，而不是個function來做這件事才對
# 而class中的__new__()，就是創建一個物件，所以如果要用class來創建類，那很可能就會和__new__()有關
# 並且，既然是要自創元類，那就不會是從object繼承下來，而是要從最基礎的元類type開始
# 元類創建類，類創建實例物件
# 可以用metaclass去做出自己的元類
# 換言之，如果看到繼承type的話，那就要特別注意，這邊的操作，很可能和元類有關
class MetaClass_upper_attr(type):
    def __new__(cls, class_name, class_parents, class_attr):
        new_attr = {}
    for name,value in class_attr.items():
        if not name.startswith("__"):
            new_attr[name.upper()] = value
    # 方法1
    return type(class_name, class_parents, new_attr)
        
    # 方法2：上面在.__new__修改，同理在type中的__new__也可以修改
    # return type.__new__(cls, class_name, class_parents, new_attr)

```

```python
《GIL - 全局解釋器鎖》
https://www.udemy.com/course/python-en/learn/lecture/15690168#overview

# 全局解釋器鎖(GIL, global interpreter lock)
# 用於同步執行緒的一種機制，使得任何時刻僅有一個執行緒在執行，即便在多核心處理器上
# 「python3 xxx.py」，那個python3其實就是個解釋器的程序，xxx.py是丟給程序的參數
# 解釋器就是，把我們寫的code，轉成機器執行的了的(2進制)code
# 使用 GIL 的解釋器只允許同一時間執行一個執行緒，常見的使用 GIL 的解釋器有CPython與Ruby MRI，像java寫的jpython就沒有GIL
# GIL不是Python的問題，而是解釋器的問題，而CPython有GIL，是個歷史問題
# 在19xx年代開發Python時，吉多寫cpython解釋器時，那時候CPU核心還沒有那麼強大，所以沒把多進程多線程那些復雜的事考慮進去
# 導致CPython中其實只有一個線程，所謂的多線程，實際上就是改成讓各線程去搶鎖，誰搶到就誰做
# 所以尷尬的是，現代已經有多個CPU，但CPython的多線程，也只能實際使用到其中一個CPU而已
# 多進程不共享全局變量，佔用資源多；多線程共享全局變量，佔用資源少
# 雖然吉多嘗試過拿掉GIL，但拿掉後衍生出的新東西，會讓CPython效率下降，所以就放棄了
```

```python
《2線程死循環.py - threading》
# 當要使用2線程，是在Python中用thread去創建的方式，而不是實際開2個terminal視窗，跑2個進程
# 那在htop會看到的，就不會是cpu1、cpu2都100%，而是pu1、cpu2都50%
# 也就是線程1在運行的時候，線程2在休息；線程2在運行的時候，線程1在休息
# 這意味着，Python創建的thread，實際上是假的多任務（並行，parallel）


import threading

# 死循環函數
def infinite_loop():
    while True:
        pass

# 創建一條子線程thread1，讓它去跑infinite_loop
thread1 = threading.Thread(target=infinite_loop)
# 在這邊讓子線程開始啓動
thread1.start()
# 這邊開始分成2個指標，一個指向主線程繼續往下走，另一個指向子線程thread1去執行infinite_loop

# 這時候主線程、子線程thread1，2條線程都進入死循環
while True:
    pass

```

```python
《2線程死循環.py - multiprocessing》
# 在Python中用多進程(multiprocess)去創建，cpu1、cpu2都會是100%
# 所以在Python中，真正可以發揮並行(parallel)，一定是多進程(multiprocess)
# 多線程就是個，假的並行(parallel)，真正的並發（complicating）
# 所以GIL在多線程裏面有，在多進程中沒有


# 把前面的多線程，改成了多進程
import multiprocessing

# 死循環函數
def infinite_loop():
    while True:
        pass

process1 = multiprocessing.Process(target=infinite_loop)
process1.start()

while True:
    pass
```

```python
《協程、線程、進程》
# 網路通訊使用多線程，會比單線程有效率
# 因爲當線程1，從Client端發送了request，在等server端回傳的時候，會進到堵塞狀態，線程1這時候就可以把鎖釋放出去給線程2
# 然後線程2也發送了request，在等server端回傳的時候，也進到堵塞狀態，又可以把鎖釋放出去給線程1
# 這時候線程1可能就已經等到server端的回傳，可以開始處理receive
# 所以線程1，可以把這段堵塞狀態的時間，給線程2活化利用，因此網路通訊使用多線程，會比單線程有效率


# 換句話說，如果主線程、子線程都是沒有堵塞時間的「計算密集型」，比如做1～1萬之間的立方和，那CPython的多線程就完全沒有一樣
# 這個例子中，不管主線程或子線程，都完全沒有堵塞時間可以去把鎖交給其他線程，所以創建線程是沒有意義的事，而是「應該用進程」
# 也就是說，堵塞時間是個重點，比如多堵塞的「I/O密集型」，那用線程就越適合，，而前面講的是網路通訊也算是input進去，等待output出來


# 協程(coroutine)，單一鎖(單核心)資源遇到堵塞時就把資源鎖交付給其他線程，只有單線所以也沒什麼共享不共享全局變量的問題
# 線程(thread)，單一鎖(單核心)資源遇到堵塞時就把資源鎖交付給其他線程，共享全局變量，無法加速單一任務，但可利用阻塞空檔並發(complicating)多任務
# 多進程(multiprocess)，多個鎖資源(多核心)，復制一份全局變量出來，可同時並行(parallel)同一項目任務，加速單一任務，也可利用阻塞空檔並發(complicating)多任務
# 多進程因爲資源不共享，因此不需要特別對進程做互斥存取的處理。但缺點是，進程間若有通訊的需求也較為複雜。
# 程式 (Program)是寫好尚未執行的 code，程式被執行後才會變成進程 (Process)


# process 是 OS 分配資源的最小單位，而 thread 則是作業系統能夠進行運算排程的最小單位
# 也就是說實際執行任務的並不是進程，而是進程中的線程，一個進程有可能有多個線程。
# 大部分的文章討論的都是 process 與 thread 的概念，直到最近golang碰到了goroutine，才知道原來還有 coroutine 協程的存在
# 進程中有線程，而線程中則又有協程，協程的排程完全由使用者控制。
# process 及 thread 的調度，是由 CPU 內核去進行調度，而協程卻不ㄧ樣，OS 甚至不知道協程的存在
# 如果 coroutine 被卡住，則會在用戶端直接切換另外一個 coroutine 給此 thread 繼續執行

```

```python
 《私有化》
# __xx__，前後雙下劃線，內建的魔法方法，會主動繼承
# __xx，前雙下劃線，除了class本身的內部當然可以訪問(不然還有誰更親密)，外部如實例、子類都看不到
# _xx，前單下劃線，實例、子類可以訪問，但更外部的from xxx import xxx，就不會導入了，所以訪問不了
# xx_，後單下劃線，避免關鍵字或重名時的解法


# 底線越多，限制就越多，比如前雙底線，後也加雙底線的，魔法方法最限制
# 其次是前雙底線，只給內部看，連生出來的實例、子類都不給看
# 再其次是前單底線，生出來的實例、子類可以看，但不給外部module導入


class Test:
    def __init__(self):
        self._v1 = 100 # 實例、子類可以訪問，但無法經由from import導入
        self.__v2 = 200 # 前雙下劃線，內部看得到，外部如實例、子類都看不到

t = Test()
print(t._v1) # 100
print(t.__v2) # 'Test' object has no attribute '__v2'
# 想一窺物件中的私有屬性，可以使用魔法方法「物件.__dict__」
# 並且可以發現，前雙底線的私有屬性名，變成了「_類名__私有屬性名」
# 所以我們就可以嘗試，使用這樣的名稱去調用前雙底線私有屬性看看（下面驗證的確是可行）
# 而前單底線的私有屬性，生出來的實例及子類是本來就可以看得到的，所以仍然保持前單底線的名稱
# 也就是說，Python解釋器，爲了讓這樣的前雙底線屬性私有化，使用的方法是，去悄悄的改了名字(名字重整)，前面多加了「_類名」
print(t._Test__v2)  # 200，調用私有屬性，物件名._類名__私有屬性名
print(t.__dict__) # {'_Test__v2': 200, '_v1': 100}


t.__aaa = 300 # 嘗試從外部加入私有屬性
# 發現在__dict__中，那個從外部加入的私有屬性名，並沒有被解釋器做名字重整
print(t.__dict__) # {'_v1': 100, '_Test__v2': 200, '__aaa': 300}
# 所以外部呼叫時，的確也會呼叫的到
# 因爲嘗試加入私有屬性的地方是外部，而呼叫的地方也是外部，所以同層級當然可以訪問
print(t.__aaa) # 300
```

```python
《import - 指向》
# Python中，本來a=100，下一行又寫a=200，這並不是把原本值100的位置，改成值200
# 而是創建了一個變量名標籤a，指向記憶體中創建的物件值100，而後又在記憶體中創建一個物件值200
# 然後把變量名標籤a，改指向值200的那個物件。讓原本指向值100的物件，減少一個指向它的指標（如果本來就只有1個指標，那就是變成沒有指標指向它）


# 當執行xxx行時，會創建一個變量名test，指向模組xxx中函式test的值(也就是地址)
# 這才使得在__main__中使用test()時，會是調用記憶體中，xxx模組的test函式
from xxx import * # module xxx中有test()
# 當執行到yyy行時，因為已經有變量名test，所以不需要再次創建
# 這時候會將變量名test，指向模組yyy中函式test的地址
# 所以當在__main__中使用test()時，會是調用記憶體中，yyy模組的test函式
# 也就是說，此時已經沒有指標指向記憶體中，xxx模組的test函式了
# 這就是後面load進來，遇到有同名的，就會覆蓋掉前面load進來的原因
# 說白了，其實也和上面寫的例子，先a=100，後面又寫了a=200的例子是一樣的道理
from yyy import * # module yyy中有同名但不同功能的test()


# 同理，如果在__main__中寫了兩次def test()，那後面的也會覆蓋掉前面的
# 這也是在Python中，Overloading重載很少出現的原因
# 像Java的話，因為參數會指定型態，不像Python這樣自由，所以Overloading比較不會造成混淆
```

```python
《import - 入坑前》
# import後面不止可以接函數、class，也可以接全局變量
# 一個程序可能由多個py模塊組成，所以被調用的module會被import進來

------------------

# 被調用的receive.py（接收msg）
from common import RECEIVE_LIST
import common

def reveive():
    print("--reveive--")
    RECEIVE_LIST = RECEIVE_LIST.append([i for i in range(3)])

def confirm_reveive():
    print("--confirm_reveive--")
    print(RECEIVE_LIST)

def reveive_next():
    print("reveive_next")
    if common.HANDLE_FLAG:
        print("--可繼續reveive--")
    else:
        print("--要暫停reveive--")

------------------

# 被調用的handle.py（處理msg）
from common import RECEIVE_LIST
import common

def handle():
    print("--handle--")
    for i in RECEIVE_LIST:
        print(i)
    common.HANDLE_FLAG = True

def confirm_handle():
    print("--confirm_handle--")
    if HANDLE_FLAG:
        print("--完成handle--")
    else:
        print("--待處理handle--")

------------------

# 存儲公用數據的common.py（比如flag等全局變量）
RECEIVE_LIST = list()
HANDLE_FLAG = False

------------------

# main.py（調用各modules）
from module1 import *
from module2 import *

def main():
    reveive() # 接收新數據
    confirm_reveive() # 確認新數據接收完成
    reveive_next() # 判斷前數據，直到處理完，才會接收新數據
    handle() # 處理新數據
    confirm_handle() # 確認新數據處理完成
    reveive_next() # 判斷前數據，直到處理完，才會接收新數據

if __name__ == "__main__":
    main()

------------------

# receive.py、handle.py，都使用import common的話，運作起來是如所預期的沒錯
！！# 把導入記憶體中common.py的「模組地址」，賦值給在receive.py、handle.py中，各別創建的變量名common
# 所以在receive.py、handle.py中，common.HANDLE_FLAG都會是common.py中的初始值，也就是Flase
# 而當handle.py執行到handle()時，有一句「common.HANDLE_FLAG = True」
# 因爲common.HANDLE_FLAG指向的是common.py，所以會把記憶體中的common.py，其中的HANDLE_FLAG改成True
# 而但receive.py的reveive_next()去做「if common.HANDLE_FLAG」判斷時，因爲指向的common.HANDLE_FLAG也是記憶體中common.py的HANDLE_FLAG，所以判斷就會得到True值
# 也就是說，在receive.py、handle.py裏面看到的common.HANDLE_FLAG，都是指向同一個位置

"""
--reveive--
--confirm_reveive--
[0, 1, 2]
--reveive_next--
--要暫停reveive--
--handle--
0
1
2
--confirm_handle--
--完成handle--
--reveive_next--
--可繼續reveive--
"""
```

```python
《import - 入坑的開始是 from import》
# 但因爲每次都要寫common.HANDLE_FLAG很麻煩，所以不想用「import common」，所以我們在handle.py改成用「from common import HANDLE_FLAG」
# 聽起來沒什麼問題，但結果就是會出乎預期
！！# 「from common import HANDLE_FLAG」，把記憶體中common.py中HANDLE_FLAG的值「False」，賦值給在handle.py中創建的變量名HANDLE_FLAG
# 所以當handle.py執行到handle()時，有一句「HANDLE_FLAG = True」時
# 因爲HANDLE_FLAG指向的是右邊的「值」，而不是左邊的變量名
# 造成handle.py的HANDLE_FLAG值改成了True，但記憶體中common.py的HANDLE_FLAG值卻沒有改變
# 因爲receive.py使用的仍然是「import common」，所以指向的common.HANDLE_FLAG是指向記憶體中common.py的HANDLE_FLAG
# 造成在receive.py和handle.py中的HANDLE_FLAG，是同名但不同值的脫鉤狀態
# 也就是不管在handle.py中怎麼去改變HANDLE_FLAG的值，都不會影響common.py的HANDLE_FLAG，和receive.py的common.HANDLE_FLAG值
```

```python
《import - 全局變量、局部變量，同名混淆》
# 此外，在handle.py中，上面有「from common import HANDLE_FLAG」的全局變量HANDLE_FLAG
# 下面有def handle()中的局部變量HANDLE_FLAG
# 當局部變量名HANDLE_FLAG設爲True時，全局變量名HANDLE_FLAG的值是不會發生改變的，2個變量同名，造成使用上的混淆
# 想到的解法是，那就把def中的局部變量名HANDLE_FLAG，加上個global前綴
# 但即使解決了，在handle.py這同一module內，指向的HANDLE_FLAG，都是全局變量，但仍然沒有解決和receive.py同名不同值的脫鉤狀態
# 所以解法是，當一個module的全局變量，是大家都會公共去取用與改變的，最好是import module就好
# 比如這個case，就是讓指向handle.py、receive.py都做import common，使用common.HANDLE_FLAG就好
# 不要想着簡化縮短成HANDLE_FLAG而已，容易自己給自己挖坑跳，還找不到原因
```

```python
《import - 等號賦值造成脫鉤》
# 如果common.py中的HANDEL_FLAG，是個list()，比如[11, 22]，而不是單一值True而已的話，會有不一樣嗎？
# 當在handle.py中使用「from common import HANDLE_FLAG」，相當於創建一個變量名HANDLE_FLAG，指向common.py中「HANDEL_FLAG=list()」右邊list()，而不是左邊的變量名HANDEL_FLAG
# 所以當在handle.py中使用「HANDLE_FLAG.append(33)」後，common.py中指向的那個list()，也會被改變成[11, 22, 33]
# 也就是說，雖然handle.py使用的是「from ... import ...」，但handle.py對HANDLE_FLAG的操作，卻會造成common.py中的HANDLE_FLAG值也會被改變
# 然而，當handle.py對HANDLE_FLAG的操作，改成等號賦值的方式，比如「HANDLE_FLAG = [11, 22, 33]時
# 就相當於將一開始handle.py中「from common import HANDLE_FLAG」所創建的變量名HANDLE_FLAG，指向了新創建的物件[11, 22, 33]
# 也就是再度的和common.py的變量HANDLE_FLAG脫鉤了
# 所以handle.py經由等號賦值對HANDLE_FLAG造成的改變，不會對receive.py有任何影響，2個py檔的變量名HANDLE_FLAG是脫鉤狀態
```

```python
《__dict__ 與 dir()的區別》
https://reurl.cc/Kbv8de

實例的__dict__，僅儲存與實例相關的屬性，正是因為如此，每個實例屬性之間才會互不影響。
類的__dict__儲存所有實例共享的變數和函式(類屬性，方法等)，
類的__dict__並不包含其父類的屬性，但dir()的返回則會包含從父類中繼承的屬性。
所以如果想獲取一個物件所有有效屬性，應使用dir()。

dir()是一個函式，返回的是list，包含類屬性、物件屬性，甚至父類屬性。
dir()有個子集__dict__，是一個字典，鍵為屬性名，值為屬性值。
並不是所有物件都擁有__dict__屬性，許多內建型別就沒有__dict__屬性，如list。
此時就需要用dir()來列出物件的所有屬性。

class A(object):
    class_var = 1

    def __init__(self):
        self.name = 'xy'
        self.age = 2

    @property
    def num(self):
        return self.age

    def fun(self):
        pass
    
    @classmethod
    def class_f(cls):
        pass

    @staticmethod
    def static_f():
        pass

if __name__ == '__main__':#主程式
    a = A()
    print(a.__dict__)  # {'name': 'xy', 'age': 2}，實例屬性
    print(A.__dict__)  # 類屬性中，包含各種方法名，也包含存放實例屬性的__dict__
    """
    {'__module__': '__main__', # 本執行處的module
    'class_var': 1, # 類屬性
    '__init__': <function A.__init__ at 0x7f2e718f2050>, # __init__是個變量名，指向的是一個A類的__init__函數
    'num': <property object at 0x7f2e718f9b30>, # 特性物件
    'fun': <function A.fun at 0x7f2e718f29e0>, # 實例方法
    'class_f': <classmethod object at 0x7f2e71909310>,  # 類方法
    'static_f': <staticmethod object at 0x7f2e719094d0>, # 靜態方法
    '__dict__': <attribute '__dict__' of 'A' objects>, # 存放物件屬性的字典
    '__weakref__': <attribute '__weakref__' of 'A' objects>, # 弱引用
    '__doc__': None} # 類的文件說明串
    """
    a.level1 = 3 # 增加物件屬性level1
    a.fun = lambda :x # 增加物件屬性fun，並以lambda的方式賦值
    print(a.__dict__) # 可以發現，a的物件屬性多了'level1'、'fun'，其中'fun'指向的是一個lambda函數
    # {'name': 'xy', 'age': 2, 'level1': 3, 'fun': <function <lambda> at 0x7f2e75bacb90>}
    print(A.__dict__) # 類屬性與上述結果相同。各別物件獨自增加的屬性，並不會顯示在類屬性的表面，要進到各別物件的__dict__才看得到
```

```python
《多態》
# Polymorphism多態，就是同樣一句函式呼叫，只要參數不同，實際執行的就可以是完全不同的方法
# Overloading重載，則是同一個函式的參數個數，或是參數型態，或是參數順序不同，例如type()就有2種完全不同用法
# Overloading可閱讀性低，很容易造成混淆，所以盡量避免
# override覆寫，子類修改父類原有的方法

class MiniOS():
    # 2. 執行pycharm.insatall()
    # 3. 但是在Pycharm類找不到insatall()方法，所以往父類找
    # 4. app.install()執行的是App類下的insatall方法
    # 6. 執行chrome.insatall()
    # 7. app.install()執行的是Chrome類下的insatall方法
    # 8. 同樣一句app.install()，實際執行的卻是完全不同的方法，這就是多態
    def os_install_app(self, app):
        app.install()

class App():
    def install():
        print("--install_App--")

class PyCharm(App):
    pass

class Chrome(App):
    # 先執行一些自己增加的改寫，然後才執行父類方法
    def install():
        print("--install_Chrome--")
        super().install()


linux = MiniOS()
pycharm = PyCharm()
chorme = Chrome()

# 1. MiniOS類的物件，呼叫方法os_install_app，並傳入實參pycharm
linux.os_install_app(pycharm)
# 5. MiniOS類的物件，呼叫方法os_install_app，並傳入實參chrome
linux.os_install_app(chrome)
```

```python
《調用父類方法 - 類名.方法(self, ...)》
# 這邊加self，是因為__init__是個實例方法
# 當使用「類名.方法(self, ...)」方式去調用的話，好處是比super()明白易懂
# 壞處是在多繼承時，可能會造成前2層以上的父類被調用多次（這邊是示範被初始化多次）


class Parent():
	def __init__(self, p):
		print("__init__parent-開始")
		self.p = p
		print("__init__parent-結束")

class Son1(Parent):
	def __init__(self, p, s1):
		print("__init__Son1-開始")
		self.s1 = s1
		Parent.__init__(self, p)
		print("__init__Son1-結束")

class Son2(Parent):
	def __init__(self, p, s2):
		print("__init__Son2-開始")
		self.s2 = s2
		Parent.__init__(self, p)
		print("__init__Son2-結束")

class Grandson(Son1, Son2):
	def __init__(self, p, s1, s2, g):
		print("__init__Grandson-開始")
		self.g = g
		# 上面有多繼承比如5個的話，類似的呼叫就要寫5次，很重複性，之後要改code也麻煩
		# 而如果是用super()的話，這邊就只要寫一次就好
		# 剩餘的4次，其實是分散在其他父類的code中有寫super()語句，才可以完成C3算法的mro調用順序
		Son1.__init__(self, p, s1)
		Son2.__init__(self, p, s2)
		print("__init__Grandson-結束")


gs = Grandson(1, 21, 22, 3) # Grandson(parent, son1, son2, grandson)
"""
__init__Grandson-開始 # 從Grandson一路衝到parent
__init__Son1-開始
__init__parent-開始 # 第1次調用parent
__init__parent-結束
__init__Son1-結束
__init__Son2-開始
__init__parent-開始 # 第2次調用parent
__init__parent-結束
__init__Son2-結束
__init__Grandson-結束
"""
```

```python
# 《調用父類方法 - __mro__》
# super(父類名, self).方法()
# 為了解決自己依循父類路徑一路往上找時，遇到派生(兄弟)類會有各種問題，所以使用super()根據MRO順序去調用
# 使用super()的話，會一路往上衝到最頂的父類，然後一路退回到入口。先進後出
# 換言之，如果單純想調用一次父類而已，卻發現使用super()沒有達到效果
# 那要嘛直接父類名.方法()，要嘛麻煩繞路又可讀性差的，去查mro表得到父類的前一個類名xx，然後super(xx, self).方法()


# __mro__是Python解釋器中的C3算法的結果，C3可以確保每個類只調用一次的算法
# 此外，__mro__，也是當遇到在當前類找不到類屬性、方法時，會去往上尋找的一個順序


# 在函數中寫的super()，就只會往mro的下一類調用一次而已
# 所以如果下一類也要繼續，那就要在下一類中被調用的地方也寫上super()，才可以延續mro
# 而在下一類中看到super時，則是拿自己的類名，去mro元組中做搜尋，匹配到後就調用自己的下一類

class Parent():
    # 為了避免後面因爲多繼承，而發生傳入實參和接收形參數量不匹配的情況
    # 所以這邊使用不定長參數去接收，反正多拿到的沒關係，存著不用就好了
    def __init__(self, p, *args, **kwargs):
        print("__init__parent-開始")
        self.p = p
        print("__init__parent-結束")

class Son1(Parent):
    def __init__(self, p, s1, *args, **kwargs):
        print("__init__Son1-開始")
        self.s1 = s1
        # Parent.__init__(self, p)
        super().__init__(p, *args, **kwargs)
        print("__init__Son1-結束")

class Son2(Parent):
    def __init__(self, p, s2, *args, **kwargs):
        print("__init__Son2-開始")
        self.s2 = s2
        # 因爲根據mro，知道要調用的會是Parent.__init__()
        # 而Parent的__init__()方法，需要的參數是(self, p, *args, **kwargs)
        # 1. 要被呼叫的參數有self，所以是個實例方法，也就是super(type類[, object物件-or或-type類])其中的
        # 「object物件-or或-type類」，在這邊會是super(type類, object物件])，也就是
        # 2. 而要呼叫Parent的，實際上的實參就只需要傳入(p, *args, **kwargs)
        # 所以1.和2.合起來就變成super(Son2, self).__init__(p, *args, **kwargs)
        # 又因爲Son2就是當前類名，所以可以super內的參數可以都省略
        super().__init__(p, *args, **kwargs)
        print("__init__Son2-結束")

class Grandson(Son1, Son2):
    def __init__(self, p, s1, s2, g):
        print("__init__Grandson-開始")
        self.g = g
        # 因為super的參數為空，所以default是當前類，也就是等同於super(Grandson, self)
        # 所以根據Grandson.__mro__，會去調用當前類的下一個類，也就是Son1.__init__
        # 換言之，如果這邊寫成super(Son2, self)，那根據mro的下一個，就會調用Parent.__init__
        super().__init__(p, s1, s2)
        print("__init__Grandson-結束")


print(Grandson.__mro__) # 類名.__mro__會回傳元組
"""
(<class '__main__.Grandson'>, 
<class '__main__.Son1'>, 
<class '__main__.Son2'>, 
<class '__main__.Parent'>, 
<class 'object'>)
"""
gs = Grandson(1, 21, 22, 3) # Grandson(parent, son1, son2, grandson)
"""
__init__Grandson-開始
__init__Son1-開始
__init__Son2-開始
__init__parent-開始
__init__parent-結束
__init__Son2-結束
__init__Son1-結束
__init__Grandson-結束
"""
```

```python
《不定參數 打包 拆包 - *args, **kwargs》
def test1(a1, b1, *args, **kwargs):
    print(a1) # 11
    print(b1) # 22
    print(args) # (33, 44, 55)，不拆包
    print(kwargs) # {'kw1': 'aa', 'kw2': 'bb'}，不拆包
    print(*args) # 33 44 55，拆包
    print(kwargs) # TypeError，字典沒辦法拆包
    
    # 因爲外部傳進來的不定參數，會打包成元組、字典
    # 所以不管內部要用的時候，要先做*拆包，才可以變回外部傳進來時，那原原本本沒有打包的樣子
    # 使用上才會是如我們所預期的
    test2(a1, b1, *args, **kwargs)
    # 11
    # 22
    # (33, 44, 55)
    # {'kw1': 'aa', 'kw2': 'bb'}


    # 如果不拆包，那test2取完固定參數後(頭兩個)，剩下的參數還有元組以及字典
    # 多餘的參數，會被test2以*args, **kwargs去接收
    # 而*args負責接收一般參數, **kwargs負責接收關鍵字參數
    # 而在test1中傳過去test2的實參，剩下的參數只剩下元組以及字典
    # 而元組以及字典，都不是關鍵字參數，所以kwargs={}
    # 只好以一般*args的元組方式去接收，然後就變成了args=(元組, 字典)
    test2(a1, b1, args, kwargs)
    # 11
    # 22
    # ((33, 44, 55), {'kw1': 'aa', 'kw2': 'bb'})
    # {}


# 下面是原本就會有的預期，所以沒什麼問題
def test2(a2, b2, *args, **kwargs):
    print(a2) # 11
    print(b2) # 22
    print(args) # (33, 44, 55)
    print(kwargs) # {'kw1': 'aa', 'kw2': 'bb'}


test1(11, 22, 33, 44, 55, kw1='aa', kw2='bb')
```

```python
《繼承不是復制 - 練習題》
class Parent():
    x =1 # 類屬性

class Child1(Parent):
    pass

class Child2(Parent):
    pass
    

# Parent.x -> 用類名去訪問類屬性，OK
# Child1、Child2中都沒有類屬性x，所以根據__mro__去往上尋找，因而也得到1
print(Parent.x, Child1.x, Child2.x) # 1 1 1
Child1.x = 2 # 爲Child1添加類屬性
# 所以只剩下Child2找不到x，因而需要根據__mro__去往上尋找，而得到1
print(Parent.x, Child1.x, Child2.x) # 1 2 1
Parent.x = 3 # 修改Parent類屬性
# 所以Child2找不到x，根據__mro__去往上尋找，也會變成得到3
# 繼承不是復制，如果有這樣的誤解，就會以爲答案是 # 3 2 1
print(Parent.x, Child1.x, Child2.x) # 3 2 3
```

```python
《property - 物件屬性》
# 當有些屬性是必須經由內部方法來生成時，這些專門用來返回生成屬性的方法，就會被掛上@property裝飾器以作識別
# 應用例子比如，一個網頁的數據如果不拆分成好幾頁，全部都要在同一個頁面，那就會加長user等待的時間，也讓server負載很大
# 所以當user點擊頁碼時，這時候就需要有方法item_start()和item_end()，從數據庫取得要展示的品項們
# 而使用page.start和page.end這樣的屬性值操作，會讓code具有比較直觀性的可閱讀性
# 所以我們就把專門用來返回生成屬性值的item_start()和item_end()方法，掛上@property裝飾器


class Page:
    def __init__(self, current_page):
        self.current_page = current_page # 當前頁碼

    @property
    def item_start(self):
        return satrt_item
    
    @property
    def item_end(self):
        return end_item


# 如果不掛@property，那page.start()也是可以用的啊
# 但一user去調用API的角度思考，page.start明顯就是個屬性，簡單！
# 但如果調用API時，是使用page.start()，就會產生一個疑問，咦，後面是還可以接什麼參數嗎？
# 所以爲了給人家調用API沒那麼多疑惑時，如果調用時不需要傳入參數，而且回傳就是個值而已，那就掛個@property吧
page = Page(3) # 建立一個可以去數據庫查詢到第3頁的物件
page.start # 頁面起始的品項目
page.end # 頁面最後的品項目


# 簡而言之，掛上property的方法，返回的就是屬性
# 外部在調用這個屬性時，會不知道屬性其實是從property生成來的，這，也是封裝的一種實現！
# 呼叫property屬性時，就是去呼叫掛上property的那個方法名
# 但因爲本質上就是當作屬性，所以會以屬性的方式去呼叫「物件名.property方法名」，而不是用方法名後多加一個()
# 既然是當作屬性，那就不該需要傳遞參數，所以掛上property的方法，不能有形參
# 雖然不能有形參，但卻必須有一個self
# 因爲外部在呼叫屬性值時，是利用「物件名.property方法名」，也就是前面那個物件名就相當於是實參，要傳遞給形參self
# 所以如果沒有property方法沒有形參self，那使用物件名的呼叫，就會遇上沒有形參self對接的，參數數量不匹配情況


class Foo:
    @property
    def prop(self):
        return 100


foo = Foo()
prop = foo.prop # 把物件foo的property屬性，賦值給變量prop
# foo.prop是property方法的返回值，也就是100
# 當使用foo.prop()，那就相當於調用100()，所以當然會被回說，這只是個整數而已，你是要call什麼啊？？
prop = foo.prop() # 'int' object is not callable


# 驗證property方法如果除了self，還給了其他參數，會發生什麼事
class Foo:
    @property
    def prop(self, value):
        # 接收實參但實際上不需要使用到的單純「return 100」都掛了，真的用到實參的「return 100*value」也當然會掛
        return 100


foo = Foo()
prop = foo.prop # prop() missing 1 required positional argument: 'value'
# 既然如此，那我們就給value看看，不過會發現
# 因爲是property裝飾器，所以「foo.prop就已經會回傳」值100了
# 所以foo.prop(100)，就相當於調用"100"這個function ，那當然會發生沒有這個function可以調用，而出錯
prop = foo.prop(100) # Error
```

```python
《property - 私有屬性、封裝》
# 下面是在sublime中打了@property會預設跑出來的東西
# 這暗示着，掛上property的回傳值，常會是私有屬性
# 而私有屬性的特色就是，不要給外部摸到，也就是要保護起來，避免隨便屬性就被修改，或甚至刪除
# 如果開放要給改值，那就設置setter方法，要開放給刪除，那就再設置deleter方法
# 定義私有屬性，然後搭配2個公開方法方法，在其他語言也有這樣套路
# 總之，property就是對屬性值的一種封裝方式
# 而封裝，就是包起來做區隔，一方面提高避免被修改的安全性，，一方面也提高了可閱讀性

@property
def foo(self):
    return self._foo
```

```python
# 《property - setter、deleter》
# 因爲封裝的關系，所以user在調用API時，會不知道這個屬性是從property生成出來的
# 他們會覺得，既然我可以foo.prop這樣取值，那的確是個屬性沒錯
# 既然是屬性，那也就可能可以foo.prop = 12這樣的賦值，或是del foo.prop這樣刪除
# 然而，因爲foo.prop的背後其實是個property裝飾器的「方法」，所以是沒辦法執行foo.prop = 12的賦值，或是del foo.prop的刪除
# 所以我們在做給人家調用的API時，如果想開放@property屬性的賦值、刪除，就要再多做@方法名.setter、@方法名.deleter這2個裝飾器


class Goods: # python3默認繼承object，就算不寫()也一樣
    # 1. 物件屬性值，應該要定義在__init__(self)當中才比較好
    def __init__(self):
        self._price = 0
    
    # Python2沒有@xxx.setter、@xxx.deleter
    @property
    def price(self):
        print("@property")
        # 2. 如果把物件屬性值定義在這：self._price = 0
        return self._price

    # 思考一下，如果不使用@xxx.setter，而是直接def set_price(self, value)
    # 那調用Goods這個API的user，他還必須知道，有set_price(self, value)這樣的一個方法名與設置的參數才行
    # 但如果使用了@xxx.setter，他只需要知道price是個屬性就行了，要改值？那簡單，直接goods.price = 80 賦值就行了
    # 封裝好的屬性，user根本就不需要去知道背後是怎麼做出來的多餘資訊
    @price.setter
    def price(self, value): # 要給屬性賦值，所以當然必須要有一個形參來接收值
        print("@price.setter")
        # 7. 可以在這邊做驗證，比如數值範圍是否合理，user是否通過驗證，才決定是不是給予賦值
        self._price = value
        return self._price

    @price.deleter
    def price(self):
        print("@price.deleter")
        del self._price
        print("price has deleted")
        

goods = Goods()
print(goods.price) # @property，0
goods.price = 80 
print(goods.price) # @price.setter，@property，80
# 3. 那這邊就算賦值80,卻仍然會顯示是0
# 4. 因爲在「觸發@price.setter後，會再去觸發@property」，因而再次受到「初始化語句」self._price = 0的影響
# 5. 所以初始化語句，最好還是放在__init__(self)，才不會造成偵錯的困擾
# 6. setter方法，等於是多加了一個中間層保護
del goods.price # @price.deleter，price has deleted
```

```python
《open、with》
# 基礎版
# Python中只要執行到某一行遇到錯誤，process就會從那行立刻退出
# 這意味着如果在「...」那行發生了問題，那這個「文件描述符f」就沒有被關閉，而我們應該也不會注意到
# 文件描述符在process中是有數量限制的，沒關閉就會被認爲有被使用
# 那數量不夠，就會造成沒有多餘的文件描述符給我們開啓其他文件
def m1():
    f = open("output.txt", "w")
    f.write(...)
    f.close


# 升級版1
# 不管有沒有被IOError捕捉到，都一定會做finally的關閉文件動作
def m2():
    f = open("output.txt", "w")
    try:
        f.write(...)
    except IOError: 
        # 發生IOError例外的話，會執行這邊的code
    except Exception:
         # 發生其他例外的話，會執行這邊的code
    else:
        print("no exception！")
        # 沒有發生例外的話，會執行這邊的code
    finally:
        f.close()
    f.close


# 升級版2
# with就相當於try-finally的功能，所以就算發生錯誤，也會自動調用f.close()
def m3():
    with open("output.txt", "w") as f:
        f.write(...)
```

```python
《上下文管理器(context) - with》
# 上下文管理器(context)，就是實現了__enter__()和__exit__()方法的物件
# with關鍵字後面必須接上下文管理器的物件。也就是說，with會去檢查後面的物件，是否包含了__enter__()和__exit__()方法
# 延伸應用是，既然加了__enter__()和__exit__()就可以自動關閉
# 那所有需要有開就有關的東西，比如TCP/UDP的通訊協定，就也可以加入這兩個方法自動確保有好好關閉
# 如果沒有保證關閉這些開口，一方面是，在系統沒有重新啓動的情況下，會被認定這些ports是堵塞狀態，所以不能使用
# 另一方面是，就算我們資源很多，不怕這些ports堵塞，但卻可能因爲開放狀態，而遭受資安攻擊


class File1():
    # context一定要有的2個屬性，所以在初始化的地方先做好
    def __init__(self, filename, mode):
        print("--init方法--")
        self.filename = filename
        self.mode = mode

    # enter進入，就是open打開
    # 因爲要進行各種操作，所以要給一個變量名指向這個控制指標
    # 然後回傳這個控制指標給user當API，user才能執行context的操作
    def __enter__(self):
        print("--enter方法--")
        # 如果filename、mode前面忘了寫self，就會到處找不到哪裏有這兩個變量名
        self.f = open(self.filename, self.mode) 
        return self.f # 返回給「as f」的那個f

    # exit離開，就是close關起來
    # 因爲就只是要確保關閉文件而已，所以既然不需要保留什麼做回傳，當然也就不需要變量名了
    def __exit__(self, *args):
        print("--exit方法--")
        # 「這邊要關閉的，就是上面那個」__enter__所說的控制指標self.f
        # 這就是所謂的「上下文」管理器的意思，「這邊指的就是上面說的那個」
        self.f.close()


# 創建File1類的物件，傳入filename、mode的實參
# 「with context類 as xx」語句，會調用context類的__init__()
# 而又因爲as xx是user要用的操作符，所以必須先有__enter__方法回傳的控制指標self.f
# 而並非直到有讀寫等動作的f.write才去調用__enter__()方法
# 在發現要離開with這段巢式code時，不管是因爲完成了，還是error，都會自動調用__exit__()方法
with File1('output.txt', 'w') as f1:
    # --init方法--
    # --enter方法--
    print("--創建File1類的物件--")
    f1.write('aeoij;oi')
    # --exit方法--
```

```python
《上下文管理器 - @contextmanager》
# 定義類太麻煩，有@contextmanager裝飾器可以直接用
# 被裝飾的函數，會以yield分割成2部分
# yield上面的部分，會去執行__enter__()
# with context_func as f，會執行到以上的部分,而得到函數的返回值f
# yield下面的部分，則會去執行__exit__()

from contextlib import contextmanager

@contextmanager
def context_func(path, mode):
    # II. 「with xxx as f」會觸發執行__enter__()
    # 被掛載@contextmanager的函數，其yield前半部，就是__enter__()
    print('before open') # 1
    f = open(path, mode)
    print('open -> yield') # 2
    # III. 看到yield，所以__enter__()結束，回傳控制指標f，給呼叫方「with xxx as f」
    yield f
    # VI. 被掛載@contextmanager的函數，其yield後半部，就是__exit__()
    print('yield -> close') # 5
    f.close()
    print('after f.close') # 6

# I. with會先判別後面的物件，是不是有__enter__()、__exit__()
# 雖然my_open()，只是個函數，不是類，但函數掛的裝飾器裏面是有完成那2個方法，所以OK
with context_func('out.txt', 'w') as f:
    # IV. 結束__enter__()，開始使用f操作
    print('with -> f.write') # 3
    f.write('sidofjo')
    print('f.write -> end') # 4
    # V. 不管是with語句結束，或是程式出錯要離開，都會調用__exit__()
"""
before open
open -> yield
with -> f.write
f.write -> end
yield -> close
after f.close
"""
```
