```python
《Thread - 改寫run()》
# 當target=函數名，這個子線程很大，放進一個函數會太擁擠，需要使用到一個類的時候
# 讓那個類去繼承threading.Thread就好了，然後對那個類調用start()
# 因爲創建子線程就2件事，一是建立物件，二是調用start()，所以可以從只修改這2件事，來做起頭的思考點
# 既然是繼承Thread，那不改__new__、__init___的話，應該一樣可以直接類名()來創建Thread_xx物件，所以建立物件這件事就不改了
# 至於修改start()的話，感覺起來要嘛是直接覆寫start()，不然就是修改run方法、call方法之類的
# 實際上start()會去調用run方法，所以就是修改run()方法沒錯


# 在我們的Thread_xx類裏面沒有定義start()，然而Thread_xx()卻可以調用
# 思考一下就知道，那必定是父類有start()，而我們沒有去做覆寫的動作，所以自然看不到這個start方法

class Thread_xx(threading.Thread):
    # 相當於子線程的main()

    def run(self):
        self.xx1()
    
    # 相當於子線程的實例方法
    def xx1(self):
        pass

# 原本知道的是Thread(target=函數名)
# 這邊則是自定義一個類，繼承自Thread，然後改寫run()
# 而這個run()中的code，地位就相當於是子線程中的main()，可以去調用類中的各種屬性、方法
# 所以當run()結束的時候，也就相當於是這個子線程的main()結束了
t = Thread_xx()
# 這邊不是調用run()！思考一下，如果因爲自定義創建類來改寫，就要改成調用run()的話
# 那一眼看過去「t = 類名()；t.run()」，沒看到start()，不覺得這是個子線程的code啊！
# 所以在改用run()會造成可讀性變差的情況下，API給我們的使用方法是，要修改的，就封裝在類裏面
# 外部的使用呢，就和原本習慣的一樣，是「t = 類名()；t.start()」
# 此外，如果真的寫t.run()的話，那就不會是創子線程了，而是個在主線程調用一般的實例方法而已
# 父類中的start()才會去真正的創建出子線程，而要創建一個線程應該是很復雜的，不應該可以隨便去修改
# 就算真的要去修改，至少也要像去改寫，在記憶體創建物件的__new__()一樣，有super()吧！
# 而這邊的run()，連個super也沒有，是個隨便自己愛怎麼寫的子線程中的main函數
t.start()
```

```python
《坑 - append沒有改變指向，不需要global》

num = 100
ls = [11, 22]

def modify_num():
    global num

    # int、str、tuple物件的數據是不可修改
    # 所以num = num + 100會使得右邊創建一個200的Int物件，然後讓num從指向Int 100，改去指向Int 200
    # 如果要修改的是全局變量num的指向，那就必須使用global num
    num += 100

def modify_ls():
    # list是數據可修改的資料格式
    # append方法是對原本list的記憶體位置做修改
    # 全局變量ls的指向並沒有改變，所以不需要加global，就可以對全局變量ls的值做修改
    ls.append(33)
    # ls = ls + [44, 55]，兩個列表相加的結果，會創建一個新的記憶體位置來儲存，讓ls重新指向
    # 因爲要對全局變量的ls做指向改變，所以這時就必須寫global ls，不然就會error
    ls += [44, 55] # UnboundLocalError: local variable 'ls' referenced before assignment


print(num) # 100
print(ls) # [11, 22]

modify_num()
modify_ls()

print(num) # 200
print(ls) # [11, 22, 33]
```

```python
《Thread競爭關系 - 全局變量》
# 寫入的thread只有一個，但讀取有很多個，那就沒問題。怕的是，有多個寫入的thread，那就會出問題了
# Thread是共享全局變量的，process才會copy一份出來自己用
# 有共享資源需求，比如I/O密集型，才適合使用多線程
# 舉例來說，request、receive是有先後順序的，receive必須知道request結束了，才會開始動作
# 如果主線程、子線程都是沒有堵塞時間的「計算密集型」，是可以各自獨立先作業，最後再整合的，那就用進程，比如做1～1萬之間的立方和
# 但是如果兩個線程之間，不是等待關系，而是競爭關系，比如都是寫入到同一個變量，那就會出問題了
# 比如讓兩個線程都對同一個變量+1，然後各做100次，結果會是200沒錯；但當各做100萬次的時候，結果就會遠少於200萬
# 對全局變量值+1的動作大致可以拆解爲3個步驟
# 1. 獲取的變量num數值爲0 -> 2.對值做+1 -> 3.讓變量名num指向數值1
# 當線程A做完1、2步驟後，就被系統停止，將資源交付到線程B手上
# 因爲A還沒改變變量num的指向，所以此時線程B拿到的值仍然是0
# 而當線程B做完了2輪，已經讓變量num指向數值2的時候，OS將資源轉交付給A
# 此時A終於可以完成第3步驟的儲存，讓變量num指向1
# 所以狀況是，明明線程A和B總共加了3次，但是數值卻是1而已，這就是競爭關系造成的問題
# 因爲觀察到的發生次數=機率×總次數，所以當總次數變成100萬時，觀察到的次數就會變得很多了
# 歸咎原因，就是如果上面case的線程A，有讓它把加好的值存進去給變數num，也就是一套流程做完，才把資源交出去，那就沒問題了
# 資料庫也有類似這樣的操作，如果A轉賬給B的過程中沒有所有步驟都做完的話，就會把整個操作rollback，以免B還沒增加到錢，A卻被扣了錢
# 這樣的操作稱爲「原子性」，一套操作是不可分割的，要嘛全做完，要嘛全不要做
# 爲了控制多線程同步不出錯，可以對一個共享數據使用互斥鎖
```

```python
import threading
import time

# add1()、add2()都對全局變量glo_num的值一直加上去
# 因爲add1()、add2()是可獨立分開再整合的計算密集型，所以適合multi-process
# 雖然不是I/O關系，所以不適用多線程，使用共享變數會造成混亂
# 但如果就一定是要使用多線程，那就只好加上mutex互斥鎖，讓2個線程可以保持原子性
# 也就是借着一整套流程，從取值到存值完整做完，才可以把mutex互斥鎖交給別人
glo_num = 0
# mutex -> [计算机]互斥(体)
# 創建鎖，默認是解鎖狀態
# 未上鎖狀態，則會鎖定，不給別人用，只自己用而已
# 上鎖狀態，則會堵塞，自己不能用，直到這個鎖被解開才能去搶
# 這邊的互斥鎖是去控制add1()、add2()，所以必須在add1()、add2()的外層，所以也是一個全局變量
mutex = threading.Lock()


# 利用add1()做num次的加法
def add1(num): 
    global glo_num

    # 要完成一整個for，才能解鎖。如果for要執行很10秒，那user就會有感，別的線程堵塞在那10秒沒作用
    mutex.acquire()
    for i in range(num):
        glo_num += 1
    mutex.release() # 釋放鎖
    # 看子線程1做完後的結果
    print("--add1() glo_num=%d--" %glo_num)


# 和add1()一樣，沒有區別
def add2(num): 
    global glo_num

    for i in range(num):
        # 上鎖的地方越少越好，只對必要的地方最好
        # 因爲如果一個鎖的區域很大，那要等一個地方做完，可能就花費很長的時間，以至於其他線程的資源完全被壓縮了
        # 而這邊重要的是，取值、加值、存值，這3步驟的流程不能被中斷，而這一個流程其實就只是一句「glo_num += 1」的code而已
        # 所以只要對這一個流程的上下做加鎖、解鎖就好了
        # 所以最小範圍，只要加1次，只花了1微秒，就可以釋放出鎖，然後大家再去搶，所以user就無感，以爲所有線程都一起做
        mutex.acquire()
        glo_num += 1
        mutex.release()
    # 看子線程2做完後的結果
    print("--add2() glo_num=%d--" %glo_num)


# 和add2()一樣，互斥鎖最小化，只鎖定「glo_num += 1」
def add3(num): 
    global glo_num

    for i in range(num):
        mutex.acquire()
        glo_num += 1
        mutex.release()
    print("--add3() glo_num=%d--" %glo_num)


def main():
    # 創建Thread物件，然後調用start的時候才會真的創建出子線程
    # Python的主線程如果先做完，會等待子線程做完，才去把主線程結束掉
    # t1 = threading.Thread(target=add1, args=(1000000,))
    t2 = threading.Thread(target=add2, args=(1000000,))
    t3 = threading.Thread(target=add3, args=(1000000,))

    # t1.start()
    t2.start()
    t3.start()

    time.sleep(3) # 給2個子線程都能確保執行完的時間，才來看最後結果
    print("--main() glo_num=%d--" %glo_num)


if __name__ == '__main__':
    main()
    # 如果兩個線程都最小化，只鎖定「glo_num += 1」
    # 可能add2做了2次，glo_num從0變成2
    # 可能add2做了3次，glo_num從2變成5
    # add2做完了自己的100萬次，glo_num輸出1921069
    # add3沒人搶鎖了，所以持續性的把自己剩下的次數做完，輸出2000000
    # 經過3秒的sleep，確定t2、t3在這3秒一定已經都做完了，才print出最後的全局變量，現在值到底是多少
    # --add2() glo_num=1921069--
    # --add3() glo_num=2000000--
    # --main() glo_num=2000000--
```

```python
《死鎖》
# 有2個以上互斥鎖的時候，可能會遇到物件1先把資源A鎖起來，物件2也先把資源B鎖起來
# 然後物件1執行到某個階段時，要去使用到資源B，所以要去對資源B加鎖
# 然而此時的資源B，因爲早已經被物件2給上了鎖，所以物件1拿不到資源B，就變爲阻塞，開始等待物件2把資源B解鎖，才能去搶鎖繼續做下去
# 同理，當物件2執行到某個階段時，需要要去使用資源A，所以要去對資源A做加鎖的動作
# 然而此時的資源A，因爲早已經被物件1給上了鎖，所以物件2拿不到資源A，就變爲阻塞，開始等待物件1把資源A解鎖，才能去搶鎖繼續做下去
# 解決辦法有2，一是如果堵塞時間超過限制就停止process，二是銀行家算法
```

```python
import threading
import time

class Thread1(threading.Thread):
    def run(self):
        mutexA.acquire() 
        print("--Thread1 mutexA.acquire--")
        time.sleep(1) # 故意等待Thread2把mutexB加鎖
        mutexB.acquire() # 造成物件1阻塞，無法執行這步的死鎖情況
        print("--Thread1 mutexB.acquire--")
        mutexB.release() 
        mutexA.release() 


class Thread2(threading.Thread):
    def run(self):
        mutexB.acquire()
        print("--Thread2 mutexB.acquire--")
        time.sleep(1) # 故意等待Thread1把mutexA加鎖
        mutexA.acquire() # 造成物件2阻塞，無法執行這步的死鎖情況
        print("--Thread2 mutexA.acquire--")
        mutexA.release() 
        mutexB.release() 


# 創建互斥鎖物件
mutexA = threading.Lock() 
mutexB = threading.Lock()


if __name__ == '__main__':
    # 在繼承threading.Thread的類中改寫run()，這樣創建的物件使用方法start的時候，就會主動去創建子線程來調用run方法
    t1 = Thread1()
    t2 = Thread2()
    t1.start()
    t2.start()
    # --Thread1 mutexA.acquire--
    # --Thread2 mutexB.acquire--
    # 可以看到，「--Thread1 mutexB」、「--Thread2 mutexA」都沒有被印出來，因爲卡住了

```

```python
《銀行家算法 - 只有10單位，但要借20單位出去》
# 如果只有10單位，但要借20單位出去，方法是「銀行家算法」，也就是依照資金需求大小，會分1～3期數做融資
# 並且如果對其中一個客戶，有任何一筆融資明顯突然變少小或甚至爲0
# 可能會對該客戶造成倒閉的衝擊，因此發生無法償還銀行資金。既然收不回借貸出去的現金，銀行實質持有現金就會減少，而降低放貸能力
# 並且風聲會傳到其他客戶耳中，對銀行產生的不信任感，造成客戶羣陸續到銀行擠兌現金，曝露出銀行持有現金不足的問題
# 所以銀行的要點是，先以契約型式說會分期發放，實際上是從其他客戶未來回收的錢
# 衍生出的另一個要點就是，發放的金額幅度必須長期維持
A要向銀行借3單位 -> 發放3(清0) 
B要向銀行借8單位 -> 發放4(還要4) -> 發放4(清0)
C要向銀行借9單位 -> 發放2(還要7)  -> 發放2(還要5)    -> 發放5(清0)
銀行有10單位資金 -> 餘額1(賺A利息) -> 餘額8(賺AB利息) -> 餘額10(賺ABC利息)
# 把銀行家(定量的CPU資源)算法概念，應用到互斥鎖(資金需求超量的客戶)的話
# 就是對多個鎖進行分階段式的解鎖(清0)；A的解鎖(有更多資金可借貸出去)會導致B的解鎖

```

```python
《multi-process》
# exe是程序，雙擊運行起來是進程(也就是「windows管理器」中，「應用程序」右側的那欄)
# 如果同一份程式雙擊3次，那就可能創建出3個進程(如果這個程式允許的話啦)
# 而進程之間是全局變量不共享的，所以比如word等文件，才可以各自獨立不受幹擾
# 可以打「ps -aux」去看目前存在幾個同樣的程式名，就知道這個程式名被創建出幾次了
# 數據會變所以不共享，code是不變所以共享，所以子進程復制的是數據。
# 就像物件中都會有__class__，可以找到每個物件都共享的不變數據比如的類屬性
# 而不需要讓每個物件都有獨自多浪費空間擁有一份一樣的數據
# 「寫時拷貝」，有要做修改的時候，才去拷貝出來一份
# process包含「獨立數據與thread」，thread則僅僅有「共享的數據」
# 也就是說multi-process本質上仍然是靠multi-thread，只是把原本「共享的數據」多復制出來，變成「獨立的數據」
# 可以想像成，multi-thread是一份code(共享數據)上有多個箭頭在移動，multi-process是多份code(獨立數據)上各有1個箭頭在移動
# 一個PID會擁有自己的記憶體使用量、CPU使用率
# process是資源分配的單位(一個process就吃掉一個PID)，thread是OS調度的單位(在一個PID裏面再去做切分調度)
# 所以multi-process雖然比multi-thread好用的多，但缺點就是資源消耗的多
# 所以如何把不必要的multi-process使用，改爲multi-thread來做，才是需要學習的
# 一條流水線上的作業員(thread)，是去對共同的產品(數據)做修改
# 開2條流水線，等於多一倍的資源，是復制一份出來，與前一條流水線的產品(數據)獨立
```

```python
《數據共享 - put、get》
# 如果下載音樂線程A、聽音樂是線程B，那因爲數據共享，所以A下載到數據共享的地方，B就可以去取用，而播放音樂
# 那如果用的是multi-process，因爲不同process，所以數據獨立，那B要怎麼去取用A的數據，來播放音樂呢？
# 所以新的問題是，不同的process之間，數據要如何溝通呢？
# 可以使用一個中間站，比如把資料寫到硬碟或一個雲端位置
# 不過process是跑在超快記憶體，而寫到硬碟或傳到雲端是超慢的，所以雖然方法可行，但速度會比較慢
# 或是像socket通信，讓processA的對外port，去和processB的對外port做溝通
# 隊列queue是先進先出。栈stack則是先進後出，可以想像人坐進電梯，先進去的被後面進來的卡在最裏面，要等後面的人都出去了才出的出
# 所以線程A可以下載音樂到記憶體，然後再由線程B去以隊列先進先出的方式讀取音樂
# 當讀取的線程B的速度，大於下載音樂的線程A，user的畫面就會顯示數據緩衝中，也就是lag的意思

import multiprocessing

# 創建queue物件，最多可put 3條進去，不寫參數的話，會根據硬體自己決定
# queue裏面放的type可以任意，就像list的元素也可以任意一樣
q = multiprocessing.Queue(3) 
q.put("111")
q.put(222)
print(q.full()) # False
q.put([11, 22, 33])
print(q.full()) # True
# q.put(44) # 如果放超過了，就會卡在那邊，等有空位才進去


q.get() # "111"
q.get() # 222
q.get() # [11, 22, 33]


# 如果取超過
# q.get() # 空了，所以會在這邊卡住不執行，等待有東西put進來才會執行完這句
q.get_nowait() # 沒有的話就會顯示「Empty: 」，而不是卡在那邊空等

```

```python
《數據共享 - 音樂播放範例》
import multiprocessing


def download_music(q):
    # 假設下載好音樂到列表變量music去
    music = [i**2 for i in range(3,8)]
    # 因爲要以隊列的方式，和另一條子process共用數據
    # 所以要先在子process外部創建好隊列，也就是main()或其他全局變量的地方
    # 同一臺電腦就用隊列去記憶體裏面抓，不同臺電腦就用快速存取的redis

    # 要把下載的音樂寫入隊列
    for m in music:
        q.put(m)
    print("--下載的音樂全部寫入隊列了--")


def play_music(q): # 隊列q是共享的數據位置
    m_list = list() # 待播放的清單
    # 開始從隊列一條條的獲取音樂，直到empty
    while True:
        m = q.get()
        m_list.append(m)

        if q.empty():
            break

    print("--播放音樂--: ", m_list)


def main():
    # 創建隊列給兩個子線程當作共用數據的地方
    q = multiprocessing.Queue()

    # 創建2個子線程，並把共用數據的隊列當作參數
    p1 = multiprocessing.Process(target=download_music, args=(q,))
    p2 = multiprocessing.Process(target=play_music, args=(q,))
    p1.start()
    p2.start()


if __name__ == '__main__':
    main()
    # --下載的音樂全部寫入隊列了--
    # --播放音樂--:  [9, 16, 25, 36, 49]
```

```python
《進程池》
# process的創建和銷毀，是需要大量資源的。比如去和OS協調記憶體位置，編列資源空間等
# 所以我們可以創建一個進程池，比如有10個進程待命
# 當有100個任務來的時候，就丟到這10個進程，有任務結束的時候，就會有下一個任務填補到這個進程去執行
# 進程池容納了很多條進程，會重復利用裏面的process
# 池的意思就是先存着，當緩衝
# 雖然多進程可以增速，但過多會對操作系統的資源造成壓力
# 多線程也是同理，過多的情況下，那每個thread要等到被輪到的時間也就變長了
# 當任務是固定的數量，就創建固定數量的進程數(有個上限，不可能爲了10萬個任務，創建10個進程)
# 當任務是未知的數量，就使用進程池
# 比如拷貝文件，就很適合用multi-process，而一個資料夾的文件數可能很大，所以就可以創建10個進程的進程池，去循環拷貝文件
# Pool(n)會創建n條進程，但是事實，會是當真正有任務進來的時候，才會創建出來，而不是執行完Pool(n)後就創建出來
# 進程池不像multi-process、multi-thread一樣，主進程是不會去等待進程池的任務都完成的
```

```python
from multiprocessing import Pool
import os, time, random


def proceeding(process_n):
    t_start = time.time()
    # 進程池裏面如果產生異常是不會報錯出來的！
    # 比如這邊如果打os.getpid，忘記那個()，不會報錯，只是不執行進程池，所以只顯示「--program start--、--program stop--」
    # 可以用pycharm執行後，看右上角跳出來的三角形警告，查看哪裏有問題，就會發現os.getpid被反白標示起來，而猜到是忘了寫()
    # 或是去設定幾個斷點位置，用debug模式查看哪裏出錯
    print("任務%s開始執行，PID爲%d" % (process_n, os.getpid()))
    # 模擬執行時間，又怕0～1之間的隨機數太小，所以乘2倍
    time.sleep(random.random() * 2)
    t_stop = time.time()
    print("任務%s執行完成，耗費%0.2f秒" % (process_n, t_stop - t_start))

def main():
    # 創建進程池，最多可有3個進程數，但此時還沒有真正創建出進程
    pol = Pool(3)
    # 投入5個任務到這個進程池
    # 此時有任務了，才會真正的去創建進程
    # 3個任務會先把進程池塞滿，剩下的任務，會等到有空的進程，一個個搶着進去使用
    for i in range(0, 5):
        # 進程池物件.apply_async(函數名, (送去給函數的元組參數))
        pol.apply_async(proceeding, (i,))

    print("--program start--")
    pol.close()  # 關閉進程池的大門，不讓新的任務繼續進來
    pol.join()  # 要讓主進程去等待進程池的任務都完成
    # 先close關門，再做join等待完成，是有固定順序的
    print("--program stop--")

if __name__ == "__main__":
    main()
    """
    任務0開始執行，PID爲372    # 持續的把任務塞滿進程池
    任務1開始執行，PID爲373
    任務2開始執行，PID爲374 
    --program start--       # 主進程繼續進行下去
    任務2執行完成，耗費0.64秒
    任務3開始執行，PID爲374    # 有任務結束後，就可以有新的任務進去進程池執行
    任務1執行完成，耗費0.74秒
    任務4開始執行，PID爲373
    任務0執行完成，耗費0.77秒
    任務3執行完成，耗費1.65秒
    任務4執行完成，耗費1.96秒
    --program stop--        # 因爲有寫「進程池物件.join()」，所以主進程會等進程池的任務結束，才把最後一句給執行完，然後結束主進程
    """
```

```python
《進程池 - 資料夾復制》

import os, time, random
import multiprocessing


def copy_file(old_f, new_f, file, que):
    # print("開始復制%s" %(file))

    old_f_rb = open(old_f + "/" + file, "rb")
    content = old_f_rb.read()
    old_f_rb.close()

    time.sleep(random.random() * 3)

    new_f_wb = open(new_f + "/" + file, "wb")
    new_f_wb.write(content)
    new_f_wb.close()

    # print("完成復制%s" %(file))
    que.put(file) # 放入隊列，讓主進程知道哪個file完成了復制


def main():
    # old_f = input("輸入要copy的文件夾名稱")
    old_f = "/usr/lib/python3.7/"
    
    try: # 如果已經存在復制後的資料夾，程式也不會報錯
        # new_f = old_f + "(復制)"
        new_f = "test/"
        os.mkdir(new_f)
    except:
        pass

    files = os.listdir(old_f) # 資料夾下的所有檔案名

    # 任務之間沒有等待關系，所以用multi-process
    # 任務數很多，所以用進程池
    pol = multiprocessing.Pool(5) 

    # 創建一個隊列，讓子線程可以去記錄目前完成了哪些task，然後再讓主線程去取用這些已完成的task
    # 因爲主線程閒着也是閒着，就讓它不停的去更新打印進度條好了
    # 進程用multiprocessing.Queue()
    # 進程池用multiprocessing.Manager.Queue()
    # Queue()不寫參數的話，會根據硬體自己決定
    que = multiprocessing.Manager().Queue() 

    for file in files:
        # 添加任務到進程池
        # print("添加%s復制任務到進程池" %file)
        pol.apply_async(copy_file, args=(old_f, new_f, file, que))

    pol.close() # 關閉進程池的大門，不給添加新任務了
    # pol.join() # 要主進程停在這邊，等待進程池的任務都結束

    files_num = len(files)
    task_done_num = 0
    while True: # 不停的一直做重復的事
        # Queue物件.get()會一直卡在那邊等東西進來
        # 所以一有copy完成的file，就開始做下面的事
        que.get()
        # print("主線程收到子線程完成拷貝任務%s" %(file))
        task_done_num += 1
        # 加個「\r」，是爲了把遊標移動到行頭，複寫掉上次的進度條
        # %%才會顯示一個%，如同\\才會顯示一個\
        print("\r拷貝進度：%.1f %%" %(task_done_num*100/files_num), end="")

        if task_done_num >= files_num:
            break

if __name__ == '__main__':
    main()
    # 可以看得出來，用多線程的話，打印順序會很亂，不是直觀可以找到的規律
    """
    開始復制warnings.py
    ...
    開始復制sched.py
    添加warnings.py復制任務到進程池
    ...
    添加getopt.py復制任務到進程池
    完成復制sched.py
    開始復制this.py
    主線程收到子線程完成拷貝任務getopt.py
    拷貝進度：0.5 %完成復制_collections_abc.py
    開始復制mailbox.py
    完成復制warnings.py
    開始復制wsgiref
    開始復制cProfile.py
    完成復制os.py
    主線程收到子線程完成拷貝任務getopt.py
    拷貝進度：1.0 %主線程收到子線程完成拷貝任務getopt.py
    拷貝進度：1.5 %主線程收到子線程完成拷貝任務getopt.py
    拷貝進度：2.0 %開始復制netrc.py
    完成復制this.py
    開始復制quopri.py
    完成復制mailbox.py
    開始復制collections
    開始復制xdrlib.py
    主線程收到子線程完成拷貝任務getopt.py
    拷貝進度：2.5 %主線程收到子線程完成拷貝任務getopt.py
    拷貝進度：2.9 %完成復制cgitb.py
    """
```
